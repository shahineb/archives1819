## Logistic Regression



Today we’ll be moving from linear regression to logistic regression, one of the simplest ways to deal with a classification problem. Instead of fitting a line, logistic regression models the probability that the outcome is 1 given the value of the predictor. In order to do this we need a function that transforms our predictor variable to a value between 0 and 1. Lots of functions can do that, but the logistic function is the most common choice:

$$f(z) = \frac{1}{1+\exp{-z}}.$$

To predict the class of our observations we'll have to minimize the corresponding loss function and as we are in a high-dimensional context we'll add an $l_2$ regularization to the model:

$$L(\textbf{w}) = \sum_{i=1}^n log(1+\exp(-y_i\textbf{w}^Tx_i))+\frac{\lambda}{2} \| \textbf{w} \|^2,$$

where $x_i$ is the vector of features for the observation $i$ and $y_i \in \{-1, 1\}$ is the class label.  


We first use the `sklearn` implementation:

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(penalty="l2", C=1.0, solver="liblinear") 
model.fit(w8a_train_x, w8a_train_y)
y_pred = model.predict(w8a_test_x)
```

and we compute the accuracy score to evaluate the model performance:

```python
from sklearn.metrics import accuracy_score
print("{:.2%}".format(accuracy_score(w8a_test_y, y_pred)))
```

## Question 1: Expression of the gradient and deterministic GD implementation

The function $F$ can also be written 

$$F(w) = G(w) + \frac{\lambda}{2} ||w||^2$$ 

where

$$G(w) = \sum_{i = 1}^n \log(1 + \exp(\left[ LB \right]_i)) \> \text{ and } \> L = (-y_i x_{ij})_{(i, j)} \in M_{n, p}(\mathbb{R})$$

The gradient of the function $G$ can then be written:

$$(\nabla G(w))_j = \sum_{i=1}^n -y_i x_{ij} \frac{\exp (-y_i w^T x_i)}{1 + \exp (-y_i w^T x_i)}, \> j \in \{ 1, ..., p \}$$.

Therefore, we obtain:

$$(\nabla F(w))_j = \sum_{i=1}^n -y_i x_{ij} \frac{1}{1 + \exp (+y_i w^T x_i)} + \lambda,\> j \in \{ 1,..., p \}$$

### Deterministic Gradient Descent

We will now implement the deterministic gradient descent to be compared with the more sophisticated methods.

__Initialization:__ We chose to initialize $\beta^{(0)}$ as the zero vector.

__Iteration `t` of `max_iter`__

> **(a)** Compute $\nabla L_{\mathcal{B_{t}}}(\beta^{(t-1)})$

> **(b)** Compute $\beta^{(t)} = \beta^{(t-1)} - \gamma \nabla L_{\mathcal{B_{t}}}(\beta^{(t-1)})$

```python
def grdt_iteration(f, fprime, beta, lbda, X, y, lr):
    fp = fprime(beta, lbda, X, y)
    next_beta = beta - lr * fp
    return next_beta, f(next_beta, lbda, X, y)

class GradientDescent():
    """ Class for deterministic gradient descent
    
    Attributes:
    -----------
    _coef: 1-dimensional np.array
        coefficients
    _lbda: float
        regularization parameter
    lr: float
        learning rate
    _max_iter: int
        the maximum number of iterations before stopping the descent
    """
    
    def __init__(self, lbda, max_iter, lr=1e-1):
        self._lbda = lbda
        self.lr = lr
        self._max_iter = max_iter
        
    def f_lr(self, beta, lbda, X, y):         
        """ 
        Returns the logistic loss.        
        """
        somme = np.sum(np.log(1 + np.exp(- y * X.dot(beta))))
        return somme + lbda * np.linalg.norm(X) / 2

    def fprime_lr(self, beta, lbda, X, y):   
        """ Returns the gradient of f_ls at beta.
        """
        tmp = - y / (1 + np.exp(y * X.dot(beta)))
        return X.T.dot(tmp) + lbda
        
    def fit(self, X, y, beta_start=None):
        """ Fit the data (X, y).
    
        Parameters:
        -----------
        X: (num_samples, num_features) np.array
            Design matrix
        y: (num_sampes, ) np.array
            Output vector
        
        Note:
        -----
        Updates self._coef
        """
        # We add the intercept to X
        tmp = X.copy()
        X = np.ones((X.shape[0], X.shape[1] + 1))
        X[:, 1:] = tmp
        y = y.reshape(-1, 1)
        
        if not beta_start:
            beta_start = np.zeros((X.shape[1], 1))
        beta_history = [beta_start]
        f_history = [self.f_lr(beta_start, self._lbda, X, y)]

        for iteration in range(self._max_iter):
            if iteration % 10 == 0:
                print(f"{iteration} iterations done ({self._max_iter} to do).", end="\r")
            # def grdt_iteration(f, fprime, beta, lbda, X, y, lr):
            next_beta, next_f = grdt_iteration(self.f_lr, self.fprime_lr, beta_history[-1], self._lbda, X, y, self.lr)
            beta_history.append(next_beta)
            f_history.append(next_f)
        print(f"{self._max_iter} iterations done ({self._max_iter} to do).")
        self._coef = beta_history[-1]
        self._beta_history = beta_history
        self._f_history = f_history
            
    
    def predict(self, X):
        """ Make binary predictions for data X.
    
        Parameters:
        -----------
        X: (num_samples, num_features) np.array
            Design matrix
        
        Returns:
        -----
        y_pred: (num_samples, ) np.array
            Predictions (0 or 1)
        """
        # We add the intercept to X
        tmp = X.copy()
        X = np.ones((X.shape[0], X.shape[1] + 1))
        X[:, 1:] = tmp
        
        f = X.dot(self._coef)
        y = np.zeros((X.shape[0], 1))
        y[f >= 0] = 1
        y[f < 0] = -1
        return y
    
    def plot_learning(self):
        if self._f_history is not None and self._beta_history is not None:
            print(f"Number of iterations: {len(self._beta_history) -1}")
            print(f"Last loss function: {self._f_history[-1]:.2f}")
            plt.plot(range(len(self._f_history)), self._f_history)
        else:
            print("The model doesn't appear to be trained.")
```

```python
lbda = 1
max_iter = 200
gd = GradientDescent(lbda, max_iter, lr=1e-3)
gd.fit(w8a_train_x, w8a_train_y)
print("Accuracy: {:.2%}".format(accuracy_score(w8a_test_y, gd.predict(w8a_test_x))))
gd.plot_learning()
plt.title("Deterministic gradient descent with $\\lambda = 1$ and $\\alpha = 0.001$");
```

__Conclusion:__ We have an accuracy slightly inferior to the sklearn benchmark done above. The gradient descent converges in around 100 iterations with the arguments used here ($\gamma = 0.001$ and $\lambda = 1$).

## Question 3: Stochastic Gradient Algorithm

We have shown before that the function $F$ can be written:

$$F(w) = G(w) + \frac{\lambda}{2} ||w||^2$$

where

$$G(w) = \sum_{i = 1}^n \log(1 + \exp(\left[ LB \right]_i)) \> \text{ and } \> L = (-y_i x_{ij})_{(i, j)} \in M_{n, p}(\mathbb{R})$$

We can see that $F$ can be written:

$$F(w) = \sum_{i=1}^nf_i(w)$$

where

$$f_i(w) = log(1+\exp(-y_i w^Tx_i)) + \frac{\lambda}{2n} ||w||^2, \forall w\in\mathbb{R}^{d}, \forall{i}\in\{1,...,n\}$$

Therefore the gradient descent presented before can be adapted with the following iteration routine:

__Iteration `t` over `nr_iter`__
> **(a)** Generate random batch $\mathcal{B}_{t}=\{(X_{i}, y_{i})\}$ of size `bsize`

> **(b)** Compute $L_{\mathcal{B_{t}}}(\beta^{(t-1)})$ (`f`) and $\nabla L_{\mathcal{B_{t}}}(\beta^{(t-1)})$ (`fprime`)

> **(c)** Compute $\beta^{(t)}$ out of $\beta^{(t-1)}$ following the `acceleration` strategy

> **(d)** Store $L_{\mathcal{B_{t}}}, \nabla L_{\mathcal{B_{t}}}$ and $\beta^{(t)}$ if needed

Where step **(c)** differs from an acceleration strategy to the other (acceleration strategies described here will be explored in question 5):

> - `none`: $$ \beta^{(t+1)} = \beta^{(t)} - \gamma\nabla_{\beta}{L_{\mathcal{B_{t}}}(\beta^{(t)})} $$


> - `momentum`: $$ \beta^{(t+1)} = \beta^{(t)} - \gamma\nabla_{\beta}{L_{\mathcal{B_{t}}}(\beta^{(t)})} + \theta(\beta^{(t)}-\beta^{(t-1)})$$


> - `adagrad`: If $ G_{t} = (\nabla_{\beta}L_{\mathcal{B_{i}}}(\beta^{(i)}))_{i\in\{1,...,t\}} $ we have $\beta^{(t)} = \beta^{(t - 1)} - \frac{\lambda}{\sqrt{G_{t} + \epsilon}} \nabla_{\beta}{L_{\mathcal{B_{t}}}(\beta^{(t)})}$ where we will take $\epsilon = 10^{-8}$


> - `adagrad windowed` : $ G_{t} = (\nabla_{\beta}L_{\mathcal{B_{i}}}(\beta^{(i)}))_{i\in\{t - \text{window},...,t\}} $ with $\text{window}$ a constant here taken equal to 20.


> - `RMSProp` : We define the contribution $C_0 = G_{0}$ and $\forall t \geq 1, \> C_t = 0.9 C_{t-1} + 0.1 G_t$ and compute $\beta^{(t)} = \beta^{(t - 1)} - \frac{\lambda}{\sqrt{C_{t} + \epsilon}} \nabla_{\beta}{L_{\mathcal{B_{t}}}(\beta^{(t)})}$ 

```python
class StochasticGradientDescent():
    """ Class for stochastic gradient descent:
    
    Attributes:
    -----------
    _coef: 1-dimensional np.array
        coefficients 
    _lbda: float
        regularization parameter
    lr: float
        the learning rate
    bsize: integer
        size of the mini-batch
    acceleration: string
        name of the aceleration strategy
    _max_iter: int
        the maximum number of iterations before stopping the descent
    """
    
    def __init__(self, lbda, bsize=100, max_iter=1000, lr=1e-3):
        self._coef  = None
        self._lbda = lbda
        self.lr     = lr
        self.bsize  = bsize
        self._max_iter = max_iter
        
    def f_lr(self, beta, lbda, X, y):         
        """ 
        Returns the logistic loss.        
        """
        somme = np.sum(np.log(1 + np.exp(- y * X.dot(beta))))
        return somme + lbda * np.linalg.norm(X) / 2

    def fprime_lr(self, beta, lbda, X, y):   
        """ Returns the gradient of f_ls at beta.
        """
        tmp = - y / (1 + np.exp(y * X.dot(beta)))
        return X.T.dot(tmp) + lbda
        
    def fit(self, X, y, beta_start=None):
        """ Fit the data (X, y).
    
        Parameters:
        -----------
        X: (num_samples, num_features) np.array
            Design matrix
        y: (num_sampes, ) np.array
            Output vector
        
        Note:
        -----
        Updates self._coef
        """
        # We add the intercept to X
        tmp = X.copy()
        X = np.ones((X.shape[0], X.shape[1] + 1))
        X[:, 1:] = tmp
        y = y.reshape(-1, 1)
        
        if not beta_start:
            beta_start = np.zeros((X.shape[1], 1))
        beta_history = [beta_start]
        f_history = [self.f_lr(beta_start, self._lbda, X, y)]

        for iteration in range(self._max_iter):
            if iteration % 10 == 0:
                print(f"{iteration} iterations done ({self._max_iter} to do).", end="\r")
            
            batch_index = np.random.choice(X.shape[0], self.bsize)
            X_batch = X[batch_index, :]
            y_batch = y[batch_index, :]
            
            next_beta, _ = grdt_iteration(self.f_lr, self.fprime_lr, beta_history[-1], self._lbda, 
                                               X_batch, y_batch, self.lr)
            next_f = self.f_lr(next_beta, self._lbda, X, y)
            beta_history.append(next_beta)
            f_history.append(next_f)
        print(f"{self._max_iter} iterations done ({self._max_iter} to do).")
        self._coef = beta_history[-1]
        self._beta_history = beta_history
        self._f_history = f_history
            
    
    def predict(self, X):
        """ Make binary predictions for data X.
    
        Parameters:
        -----------
        X: (num_samples, num_features) np.array
            Design matrix
        
        Returns:
        -----
        y_pred: (num_samples, ) np.array
            Predictions (0 or 1)
        """
        # We add the intercept to X
        tmp = X.copy()
        X = np.ones((X.shape[0], X.shape[1] + 1))
        X[:, 1:] = tmp
        
        f = X.dot(self._coef)
        y = np.zeros((X.shape[0], 1))
        y[f >= 0] = 1
        y[f < 0] = -1
        return y
    
    def plot_learning(self):
        if self._f_history is not None and self._beta_history is not None:
            print(f"Number of iterations: {len(self._beta_history) -1}")
            print(f"Last loss function: {self._f_history[-1]:.2f}")
            plt.plot(range(len(self._f_history) - 1), self._f_history[1:])
        else:
            print("The model doesn't appear to be trained.")
```

```python
%%time
bsize = 1000
max_iter = 200

fig, ax = plt.subplots(1, 2)

print("Stochastic Gradient Descents")

lbda = 0.5
lr = 1e-4
gd = StochasticGradientDescent(lbda, bsize, max_iter, lr=lr)
gd.fit(w8a_train_x, w8a_train_y)
print("Iteration 1 : Learning rate ({:.4f}), Accuracy: ({:.2%}), Last loss val: {:.2f}".format(
    lr,
    accuracy_score(w8a_test_y, gd.predict(w8a_test_x)),
    gd._f_history[-1]
))
ax[0].plot(range(len(gd._f_history) - 1), gd._f_history[1:], 
           color=cp[1], label="LR 0.0001", alpha=0.8)

lbda = 0.5
lr = 1e-3
gd = StochasticGradientDescent(lbda, bsize, max_iter, lr=lr)
gd.fit(w8a_train_x, w8a_train_y)
print("Iteration 2 : Learning rate ({:.4f}), Accuracy: ({:.2%}), Last loss val: {:.2f}".format(
    lr,
    accuracy_score(w8a_test_y, gd.predict(w8a_test_x)),
    gd._f_history[-1]
))
ax[0].plot(range(len(gd._f_history) - 1), gd._f_history[1:], 
           color=cp[0], label="LR 0.001", alpha=0.8)

lbda = 0.5
lr = 1e-2
gd = StochasticGradientDescent(lbda, bsize, max_iter, lr=lr)
gd.fit(w8a_train_x, w8a_train_y)
print("Iteration 3 : Learning rate ({:.4f}), Accuracy: ({:.2%}), Last loss val: {:.2f}".format(
    lr,
    accuracy_score(w8a_test_y, gd.predict(w8a_test_x)),
    gd._f_history[-1]
))
ax[0].plot(range(len(gd._f_history) - 1), gd._f_history[1:], 
           color=cp[5], label="LR 0.01", alpha=0.8)

lbda = 0.5
lr = 1e-1
gd = StochasticGradientDescent(lbda, bsize, max_iter, lr=lr)
gd.fit(w8a_train_x, w8a_train_y)
print("Iteration 4 : Learning rate ({:.4f}), Accuracy: ({:.2%}), Last loss val: {:.2f}".format(
    lr,
    accuracy_score(w8a_test_y, gd.predict(w8a_test_x)),
    gd._f_history[-1]
))
ax[0].plot(range(len(gd._f_history) - 1), gd._f_history[1:], 
           color=cp[3], label="LR 0.1", alpha=0.8)


print("\nGradient Descents")

lbda = 0.5
lr = 1e-3
gd = GradientDescent(lbda, max_iter, lr=lr)
gd.fit(w8a_train_x, w8a_train_y)
print("Iteration 1 : Learning rate ({:.4f}), Accuracy: ({:.2%}), Last loss val: {:.2f}".format(
    lr,
    accuracy_score(w8a_test_y, gd.predict(w8a_test_x)),
    gd._f_history[-1]
))
ax[1].plot(range(len(gd._f_history) - 1), gd._f_history[1:], color=cp[0], label="LR 0.001", alpha=0.8)

lbda = 0.5
lr = 1e-4
gd = GradientDescent(lbda, max_iter, lr=lr)
gd.fit(w8a_train_x, w8a_train_y)
print("Iteration 2 : Learning rate ({:.4f}), Accuracy: ({:.2%})".format(
    lr,
    accuracy_score(w8a_test_y, gd.predict(w8a_test_x)),
    gd._f_history[-1]
))
ax[1].plot(range(len(gd._f_history) - 1), gd._f_history[1:], color=cp[1], label="LR 0.0001", alpha=0.8)

lbda = 0.5
lr = 1e-2
gd = GradientDescent(lbda, max_iter, lr=lr)
gd.fit(w8a_train_x, w8a_train_y)
print("Iteration 3 : Learning rate ({:.4f}), Accuracy: ({:.2%})".format(
    lr,
    accuracy_score(w8a_test_y, gd.predict(w8a_test_x)),
    gd._f_history[-1]
))
ax[1].plot(range(len(gd._f_history) - 1), gd._f_history[1:], color=cp[5], label="LR 0.01", alpha=0.8)

ax[0].set_ylim((0, 50000))
ax[1].set_ylim((0, 50000))

ax[0].legend()
ax[0].set_title("Stochastic gradient descent")
ax[1].legend()
ax[1].set_title("Gradient descent")

plt.show()
```

## Question 5: Acceleration methods

```python
def grdt_iteration_momentum(f, fprime, beta, lbda, X, y, lr, past_betas, theta):
    fp = fprime(beta, lbda, X, y)
    next_beta = beta - lr
    if len(past_betas) > 2:
        next_beta += theta * (past_betas[-1] - past_betas[-2])
    return next_beta, f(next_beta, lbda, X, y)

def grdt_iteration_ga(f, fprime, beta, lbda, X, y, lr, past_gradients):
    fp = fprime(beta, lbda, X, y)
    if past_gradients is None:
        all_gradients = fp
    else:
        all_gradients = np.concatenate((past_gradients, fp), axis=1)
    next_beta = beta - lr * np.mean(all_gradients, axis=1).reshape((beta.shape[0], 1))
    return next_beta, f(next_beta, lbda, X, y), all_gradients

def grdt_iteration_adagrad(f, fprime, beta, lbda, X, y, lr, past_gradients):
    fp = fprime(beta, lbda, X, y)
    if past_gradients is None:
        all_gradients = fp
    else:
        all_gradients = np.concatenate((past_gradients, fp), axis=1)
    next_beta = beta - lr * fp / (np.sqrt(np.sum(np.power(all_gradients, 2), axis=1).reshape(-1, 1)) + 1e-8)
    return next_beta, f(next_beta, lbda, X, y), all_gradients

def grdt_iteration_adagrad_windowed(f, fprime, beta, lbda, X, y, lr, past_gradients, window):
    fp = fprime(beta, lbda, X, y)
    if past_gradients is None:
        all_gradients = fp
    else:
        all_gradients = np.concatenate((past_gradients, fp), axis=1)
    next_beta = beta - lr * fp / (np.sqrt(np.sum(np.power(all_gradients[:, -window:], 2), axis=1).reshape(-1, 1)) + 1e-8)
    return next_beta, f(next_beta, lbda, X, y), all_gradients

def grdt_iteration_rmsprop(f, fprime, beta, lbda, X, y, lr, past_contrib):
    fp = fprime(beta, lbda, X, y)
    if past_contrib is None:
        new_contrib = 1 - 1e-8
    else:
        new_contrib = 0.9 * past_contrib + 0.1 * np.power(fp, 2)
    next_beta = beta - lr * fp / (np.sqrt(new_contrib) + 1e-8)
    return next_beta, f(next_beta, lbda, X, y), new_contrib

class StochasticGradientDescent2():
    """ Class for stochastic gradient descent:
    
    Attributes:
    -----------
    _coef: 1-dimensional np.array
        coefficients 
    _lbda: float
        regularization parameter
    lr: float
        the learning rate
    bsize: integer
        size of the mini-batch
    acceleration: string
        name of the aceleration strategy
    _max_iter: int
        the maximum number of iterations before stopping the descent
    """
    
    def __init__(self, lbda, bsize=100, max_iter=1000, lr=1e-3, acceleration="none", momentum_theta=0.5, adadelta_window=20):
        self._coef  = None
        self._lbda = lbda
        self.lr     = lr
        self.bsize  = bsize
        self.acceleration = acceleration
        self._max_iter = max_iter
        self._momentum_theta = momentum_theta
        self._adadelta_window = adadelta_window
        
    def f_lr(self, beta, lbda, X, y):         
        """ 
        Returns the logistic loss.        
        """
        somme = np.sum(np.log(1 + np.exp(- y * X.dot(beta))))
        return somme + lbda * np.linalg.norm(X) / 2

    def fprime_lr(self, beta, lbda, X, y):   
        """ Returns the gradient of f_ls at beta.
        """
        tmp = - y / (1 + np.exp(y * X.dot(beta)))
        return X.T.dot(tmp) + lbda
        
    def fit(self, X, y, beta_start=None):
        """ Fit the data (X, y).
    
        Parameters:
        -----------
        X: (num_samples, num_features) np.array
            Design matrix
        y: (num_sampes, ) np.array
            Output vector
        
        Note:
        -----
        Updates self._coef
        """
        # We add the intercept to X
        tmp = X.copy()
        X = np.ones((X.shape[0], X.shape[1] + 1))
        X[:, 1:] = tmp
        y = y.reshape(-1, 1)
        
        if not beta_start:
            beta_start = np.zeros((X.shape[1], 1))
        beta_history = [beta_start]
        f_history = [self.f_lr(beta_start, self._lbda, X, y)]

        all_gradients = None
        adadelta_decay_contrib = None
        for iteration in range(self._max_iter):
            if iteration % 10 == 0:
                print(f"{iteration} iterations done ({self._max_iter} to do).", end="\r")
            
            batch_index = np.random.choice(X.shape[0], self.bsize)
            X_batch = X[batch_index, :]
            y_batch = y[batch_index, :]
            
            if self.acceleration == "none":
                next_beta, _ = grdt_iteration(self.f_lr, self.fprime_lr, beta_history[-1], self._lbda, 
                                                   X_batch, y_batch, self.lr)
            elif self.acceleration == "momentum":
                next_beta, _ = grdt_iteration_momentum(self.f_lr, self.fprime_lr, beta_history[-1], 
                                                            self._lbda, X_batch, y_batch, self.lr, beta_history,
                                                            self._momentum_theta
                                                           )
            elif self.acceleration == "gradient_averaging":
                next_beta, _, all_gradients = grdt_iteration_ga(
                    self.f_lr, self.fprime_lr, beta_history[-1], self._lbda, X_batch, y_batch, self.lr, all_gradients
                )
            elif self.acceleration == "adagrad":
                next_beta, _, all_gradients = grdt_iteration_adagrad(
                    self.f_lr, self.fprime_lr, beta_history[-1], self._lbda, X_batch, y_batch, self.lr, all_gradients
                )
            elif self.acceleration == "adagrad_windowed":
                next_beta, _, all_gradients = grdt_iteration_adagrad_windowed(
                    self.f_lr, self.fprime_lr, beta_history[-1], self._lbda, X_batch, y_batch, self.lr, 
                    all_gradients, self._adadelta_window
                )
            elif self.acceleration == "rmsprop":
                next_beta, _, adadelta_decay_contrib = grdt_iteration_rmsprop(
                    self.f_lr, self.fprime_lr, beta_history[-1], self._lbda, X_batch, y_batch, self.lr, 
                    adadelta_decay_contrib
                )
            else:
                raise NotImplemented
            beta_history.append(next_beta)
            next_f = self.f_lr(next_beta, self._lbda, X, y)
            f_history.append(next_f)
        print(f"{self._max_iter} iterations done ({self._max_iter} to do).")
        self._coef = beta_history[-1]
        self._beta_history = beta_history
        self._f_history = f_history
            
    
    def predict(self, X):
        """ Make binary predictions for data X.
    
        Parameters:
        -----------
        X: (num_samples, num_features) np.array
            Design matrix
        
        Returns:
        -----
        y_pred: (num_samples, ) np.array
            Predictions (0 or 1)
        """
        # We add the intercept to X
        tmp = X.copy()
        X = np.ones((X.shape[0], X.shape[1] + 1))
        X[:, 1:] = tmp
        
        f = X.dot(self._coef)
        y = np.zeros((X.shape[0], 1))
        y[f > 0] = 1
        y[f <= 0] = -1
        return y
    
    def plot_learning(self):
        if self._f_history is not None and self._beta_history is not None:
            print(f"Number of iterations: {len(self._beta_history) -1}")
            print(f"Last loss function: {self._f_history[-1]:.2f}")
            plt.plot(range(len(self._f_history) - 1), self._f_history[1:])
        else:
            print("The model doesn't appear to be trained.")
```

__Conclusion:__ It seems that our acceleration algorithms don't bring benefits when compared to the simple batch stochastic gradient descent presented before. The adagrad and adagrad windowed acceleration methods have significantly slower convergence speed as the new contribution to the learning rate introduced leads to shorter steps. The vanilla adagrad acceleration is the most striking to this regard as the contribution of the new samples computed converges toward 0, stopping the model from improving with _new insights_. We would like to underline again that stochastic gradient should limit the variance in the model even if the bias seems greater (here the data from the training and validation set most likely have been sampled using the same methodology, stochastic descents might perform better on data gathered differently and would be more resilient to changes in the underlying _true_ model).
