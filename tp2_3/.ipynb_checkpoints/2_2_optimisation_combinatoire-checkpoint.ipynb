Chargement des données

```python
filePath = "data/ProbaInterception.txt"
proba_interception = []

with open(filePath, "rb") as f:
    rows = f.readlines()
    for row in rows:
        proba_interception += [row.split()]
    
proba_interception = np.array(proba_interception)
proba_interception = proba_interception.astype(float)

print(proba_interception.shape)
pd.DataFrame(proba_interception)
```

```python
proba_interception[np.isnan(proba_interception)]=0
```

```python
graph = csr_matrix(proba_interception)
```

```python
Tcsr = csgraph.minimum_spanning_tree(graph)
```

```python
pd.DataFrame(Tcsr.toarray())
```

```python
import networkx as nx

def plot_graph(graph_mat, to_highlight=None):
    G = nx.DiGraph()
    plt.figure(figsize=(16,12))
    for i, row in enumerate(graph_mat):
        spy_i = str(i+1)
        reachable_spies = np.where(row>0)[0]
        for j in reachable_spies:
            spy_j = str(j+1)
            G.add_edges_from({(spy_i,spy_j)}, weight=round(row[j],3))
            
    edge_labels=dict([((u,v,),d['weight'])
                 for u,v,d in G.edges(data=True)])
    
    if to_highlight:
        edge_colors = ['y' if not edge in to_highlight else 'red' for edge in G.edges()]
    else:
        edge_colors = ['y' for edge in G.edges()]
    
    pos=nx.spring_layout(G)
    nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
    nx.draw(G,pos, node_size=800, node_color = 'salmon')
    nx.draw_networkx_labels(G,pos)
    plt.show()

        
```

```python
G = plot_graph(proba_interception)
```

```python
G = plot_graph(Tcsr.toarray())
```
